Loading all necessary libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Loading the dataset

```python
power_df = pd.read_csv("../data/AEP_hourly.csv")
```

The columns of the dataframe are

```python
power_df.columns
```

Checking the shape of the dataset

```python
power_df.shape
```

From above output, we can see that dataframe contains 121272 rows and 2 columns.

Let's check few entries inside the dataframe

```python
power_df.head()
```

Let's check the data types of each column.

```python
power_df.dtypes
```

From preceding output, we can see that Datetime column has a object datatype.  
Let's convert this column into Date datatype.

```python
power_df['Datetime'] = pd.to_datetime(power_df['Datetime'])
```

Again, let's check for the datatypes

```python
power_df.dtypes
```

From above output, we can see that Datetime column has been converted to datetime type.  

Lets look for missing values in columns.

```python
power_df.isnull().sum()
```

There are no missing value in the dataframe.

Let's next change the index of our dataframe to the Datetime column

```python
power_df = power_df.set_index('Datetime')
power_df.head()
```

From above output, we can see that Datetime has been set as an index for out dataframe.  

Let's sample our data from hourly to daily level. While aggregating we will use sum function.

```python
power_df = power_df.resample(rule='D').sum()
```

```python
power_df.head()
```

We can see from above output that data has been sampled to daily level from hourly level.   

Lets look for the new index of our dataframe.

```python
power_df.index
```

Adding Year, Month and Weekday columns to our dataframe using Datetime index coulumn.

```python
power_df['Year'] = power_df.index.year
power_df['Month'] = power_df.index.month
power_df['Weekday'] = power_df.index.day_name()
```

```python
power_df.head()
```

We can see that three columns have been added to our dataframe.

```python
power_df.loc['2004-10-02']
```

**Visualizing time series**

Lets generate a line plot of our dataset.

```python
sns.set(rc={'figure.figsize':(11, 4)})
ax = power_df['AEP_MW'].plot()
ax.set_ylabel('Daily Consumption (MW)')
```

In above plot x-axis represents year and y-axis represets power consumption in MW.  
Now instead of line plot, lets use dot plot to visualize the same data.

```python
ax = power_df['AEP_MW'].plot(marker = ".",alpha=0.5)
ax.set_ylabel('Daily Consumption (MW)')
```

The above output shows that the power consumption fluctuates from around 300000 to 500000 MW during 2006 to 2018.  
Above plot looks cluttered due to many datapoint. So, lets look into the single year's consumption plot.

```python
# Plot of year 2006 vs power consumption
ax = power_df.loc['2006','AEP_MW'].plot(marker = ".",alpha=0.5)
ax.set_ylabel('Daily Consumption (MW)')
```

```python
# Plot of year 2007 vs power consumption
ax = power_df.loc['2007','AEP_MW'].plot(marker = ".",alpha=0.5)
ax.set_ylabel('Daily Consumption (MW)')
```

From above two year plot, we can see some similiar trend. Consumption increases from Jan to  Feb and decline gradually till May.
After May consumption inceases reaching maximum at Aug and slowly decline till Nov. In beginning of Dec, consumption increses and declines at the end of the month.

Now, lets look into the month level consumption in a plot.

```python
# Plot of Jan 2007 vs power consumption
ax = power_df.loc['2007-1','AEP_MW'].plot(marker = ".",alpha=0.5)
ax.set_ylabel('Daily Consumption (MW)')
```

```python
# Plot of March 2007 vs power consumption
ax = power_df.loc['2007-3','AEP_MW'].plot(marker = ".",alpha=0.5)
ax.set_ylabel('Daily Consumption (MW)')
```

```python
# Plot of Aug 2007 vs power consumption
ax = power_df.loc['2007-8','AEP_MW'].plot(marker = ".",alpha=0.5)
ax.set_ylabel('Daily Consumption (MW)')
```

From above monthly plots, we can see that there are sudden dips in the consumption. It might have occured due to the weekend pattern.

Let's group our time series data according to months.

```python
sns.boxplot(data = power_df, x='Month', y="AEP_MW")
```

Above box plot shows the consumption grouped by the monthly period over the whole dataset. Although there are outliers, in general, consumption decreased from Jan to April and then increased from there till Aug. From Aug, we again see the same pattern of decrease and increase in consumption for rest of the months.

Now, let's group our dataset using day of a week

```python
sns.boxplot(data=power_df, x='Weekday', y='AEP_MW')
```

From above boxplot, after discarding outliers we can see that power consumption is low during weekend as compared to weekdays.

Let's resample the data based on weekly mean basis

```python
column = ["AEP_MW"]
power_weekly_mean = power_df[column].resample("W").mean()
ax = power_weekly_mean['AEP_MW'].plot()
ax.set_ylabel('Daily Consumption (MW)')
```

Above plot uses smoothing technique by calculating weekly average over whole dataset.
